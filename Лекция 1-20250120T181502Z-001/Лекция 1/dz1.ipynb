Найдите среднее значению по каждому из каналов. Ответ - среднее в 3 канале, округлить до целого.

```python
import cv2
import numpy as np
im = cv2.imread("big.jpg")
height, width = im.shape[:2]
print(np.mean(im[:,:,0]))
print(np.mean(im[:,:,1]))
print(np.mean(im[:,:,2]))
```

4. Создать массив numpy:

np.random.seed(7964)
A = np.random.randint(0, 8, size=(500, 500))

Сколько столбцов, в которых есть хоть одно значение 7?

```python
import numpy as np
np.random.seed(7964)
A = np.random.randint(0, 8, size=(500, 500))
B = A.max(axis=0)
unique, counts = np.unique(B, return_counts=True)
С=np.asarray((unique, counts)).T
print(C[7])
```

```python
import numpy as np
np.random.seed(7964)
A = np.random.randint(0, 8, size=(500, 500))
ans=0
q=0
for i in range(500):
    q=0
    for j in range(500):
        if A[j][i]==7:
            q=1
    ans+=q
print(ans)
```

Найдите индексы значений, которые меньше 4. Ответ - сумма сумм индексов строк и столбцов между собой.

```python
import numpy as np
np.random.seed(7964)
A = np.random.randint(0, 8, size=(500, 500))
B = np.where(A<4)
C=B[0].sum()
D=B[1].sum()
print(C+D)
```

```python
import numpy as np
np.random.seed(7964)
A = np.random.randint(0, 8, size=(500, 500))
B=[]
ans=0
for i in range(500):
    for j in range(500):
        if A[i][j]<4:
            B.append([i,j])
            ans+=i+j
print(ans)
```

Найдите индексы значений, которые меньше 4 и не меньше 2. Ответ - сумма сумм индексов строк и столбцов между собой.

```python
# import numpy as np
np.random.seed(7964)
A = np.random.randint(0, 8, size=(500, 500))
B = np.where((A<4) & (A>=2))
C=B[0].sum()
D=B[1].sum()
print(C+D)
```
